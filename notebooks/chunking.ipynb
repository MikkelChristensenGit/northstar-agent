```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader
```

```python
def load_markdown(md_path: str):
    loader = TextLoader(md_path)
    docs = loader.load()
    return docs

docs = load_markdown("/Users/mchr/Documents/pfa/data/markdown/northstar.md")
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(docs)
```

```python
import json
with open("/Users/mchr/Documents/pfa/data/chunks/chunks.jsonl", "w") as f:    
    for chunk in chunks:
        json.dump({"text": chunk.page_content, "metadata": chunk.metadata}, f)
        f.write("\n")
```
